
### 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’


```python
import requests
from bs4 import BeautifulSoup as soup
```

## Loading URL


```python
url=requests.get('https://bookpage.com/reviews')
content=url.content
```

```python
content
```

```python
soup_data=soup(content,'html.parser')
print(soup_data)
```

```python
soup_data.title.text
```

```python
Books=soup_data.find('div',{'class':"bp-well"})
```

```python
Books
```

```python
firstbook=Books.findAll('div',{'bp-block article-info'})
```

```python
firstbook[0]
```

```python
first_book=firstbook[0]
```

### Book Name

```python
first_book.h4
```

```python
first_book.h4.text.replace('\n',"")
```

### Author Name

```python
first_book.p
```

```python
first_book.p.text.replace('\n','')
```

### Genre

```python
first_book.find('p',{'class':'genre-links hidden-phone'})
```

```python
first_book.find('p',{'class':'genre-links hidden-phone'}).find('a').text
```

### Reviews

```python
reviews=first_book.find('p',{'excerpt'})
reviews
```

```python
first_book.find('p',{'excerpt'}).text.replace('\n','')
```

```python
reviews.text.replace('\n','')
```

```python
Book_Name = []
Author_Name = []
Genre = []
Review = []

for first_book in firstbook:
    Book_Name.append(first_book.h4.text.replace('\n',""))
    Author_Name.append(first_book.p.text.replace('\n',''))
    Genre.append(first_book.find('p',{'class':'genre-links hidden-phone'}).find('a').text)
    Review.append(first_book.find('p',{'excerpt'}).text.replace('\n',''))
     
    
```

```python
Book_Name
```

```python
Author_Name
```

```python
Genre
```

```python
Review
```

```python
Books_Review=list(zip(Book_Name,Author_Name,Genre, Review))
```

```python
import pandas as pd
df=pd.DataFrame(Books_Review[:5],columns=['Book_Name','Author_Name','Genre', 'Review'])
```
